```python
import pandas as pd
import numpy as np
import sys
sys.path.append("../../scripts/")
from importlib import reload
import TweetToVect2 as TW
import UserToVect as UVect
```

```python
users_df = pd.read_pickle('../../datasets/users_vectores.pkl')
```

```python
user_model = UVect.UserToVect()
```

```python
test_df = users_df.sort_values('n_tweets', ascending=False)[:50]
vectores = test_df.vector_user.values.tolist()
similitudes = user_model.pairwise_similarity_vectors(vectores, vectores)
```

```python
user_ids = test_df.user_id.values
```

```python
for i in range(len(similitudes)):
    user = test_df.loc[user_ids[i],:]
    sorted_keys = np.argsort(similitudes[i])[::-1]
    sorted_similarity = np.sort(similitudes[i])[::-1]
    compared_users = test_df.loc[user_ids,:].iloc[sorted_keys]
    description = user.description.replace('\n', ' ') if isinstance(user.description, str) else ''
    print(user['name'], description)
    for i in range(5):
        key = i+1
        description = compared_users.iloc[key].description.replace('\n', ' ') if isinstance(compared_users.iloc[key].description, str) else ''
        print('\t', round(sorted_similarity[key], 2), compared_users.iloc[key,:]['name'], description)
    print('')
```
